```python
import pandas as pd

# Selecting numerical variables
numerical_variables = ['Price','Rooms','Distance','Postcode','Bedroom2','Bathroom','Car','Landsize','BuildingArea','YearBuilt','Lattitude','Longtitude','Propertycount']

# Calculate descriptive statistics
descriptive_stats = df[numerical_variables].describe()

# Display the results
print(descriptive_stats)
```

```python
import pandas as pd

# Selecting categorical variables
categorical_variables = ['Type', 'Method','Regionname','CouncilArea']

# Display the frequency of each category
for var in categorical_variables:
    category_counts = df[var].value_counts()
    print(f'\n{var} Distribution:')
    print(category_counts)
```

```python
import matplotlib.pyplot as plt

# Selecting the categorical variable
categorical_variable = 'Type'

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(df[categorical_variable].value_counts(), labels=df[categorical_variable].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title(f'{categorical_variable} Distribution')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting numerical variables
numerical_variables = ['Rooms', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude', 'Propertycount']

# Set up subplots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 12))
fig.suptitle('Histograms of Numerical Variables', y=1.02)

# Flatten the 2D array of subplots for easier indexing
axes = axes.flatten()

# Plot histograms for each variable
for i, var in enumerate(numerical_variables):
    sns.histplot(df[var], bins=20, kde=True, ax=axes[i], color='lightgreen', edgecolor='black')
    axes[i].set_title(f'{var} Distribution')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a violin plot for the 'Price' variable
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['Price'], color='skyblue')
plt.title('Violin Plot of Price Variable')
plt.xlabel('Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting numerical variables
numerical_variables = ['Price', 'Rooms', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude', 'Propertycount']

# Calculate the correlation matrix
correlation_matrix = df[numerical_variables].corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
